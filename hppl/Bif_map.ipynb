```python
import numpy as np

import matplotlib.pyplot as plt
import sounddevice as sd

%matplotlib inline
```

```python
def PlotEvolution(r=2, x0=0.5, number_of_steps=100):
  x = np.empty(0, dtype=np.float32)
  x = np.append(x, x0)

  for n in range(1, number_of_steps):
    x_new = r*x[n-1]*(1 - x[n-1])
    x = np.append(x, x_new)

  # plt.plot(x, '.')
  # plt.plot(x)
  plt.plot(x, '.-')
```

```python
PlotEvolution(r=0.1)
```

```python
PlotEvolution(2.7, 0.5, 100)
```

```python
PlotEvolution(3.3, 0.3, 100) 
```

```python
def ReturnLastNValues(r = 2.0, x0 = 0.5, number_of_steps = 100, number_of_steps_return = 50):
    x = np.empty(0, dtype=np.float32)
    x = np.append(x, x0)
    for n in range(1, number_of_steps):
        x_new = r * x[n-1] * (1 - x[n - 1])
        x = np.append(x, x_new)

    return x[-1:-number_of_steps_return-1:-1]
```

```python
for r in np.linspace(0,4, 1000):
    x = ReturnLastNValues(r=r, x0=0.5, number_of_steps = 100, number_of_steps_return = 50)
#     print(f'r = {r} \n', x, '\n')
    plt.plot(np.ones(50)*r, x, 'k.', ms = 1)
    # 
```
